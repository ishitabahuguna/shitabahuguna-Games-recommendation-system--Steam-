```python
matrix=df.pivot_table(columns='game',index='userid',values='playtime',fill_value=0)
```

recommendation system

```python
def center(row):
    a=(row-row.mean())/(row.max()-row.min())
    return a
std=matrix.apply(center)
std
```

```python
def rec(game):
    dota = std[game]
    dota = matrix.corrwith(dota).dropna()
    gameData = df.groupby('game').agg({'playtime': [np.size, np.mean]})
    gameSim = gameData['playtime']['size'] >= 100
    dfnew= gameData[gameSim].join(pd.DataFrame(dota, columns=['similarity']))
    return dfnew.sort_values(by='similarity', ascending=False)[:5]
```
